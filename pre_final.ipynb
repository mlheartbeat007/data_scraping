```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
raw_data=requests.get("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
def main():
    def find_the_all_price():
        price_of_estate=[]
        
        c=raw_data.content
        soup1=BeautifulSoup(c,'html.parser')
        all_data1=soup1.find_all('div',{'class','propertyRow'})
        #print all_data[0]
        for item in all_data1:
            price=item.find('h4',{'class':'propPrice'}).text
            price=price.replace('\n','')
            price=price.encode('ascii', 'ignore')
            price_of_estate.append(price)
        return price_of_estate
    def find_the_address():
        #raw_data=requests.get("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
        c=raw_data.content
        soup2=BeautifulSoup(c,'html.parser')
        all_data2=soup2.find_all('span',{'class','propAddressCollapse'})
        add=[]
        for item in all_data2:
            dd=item.text
            dd=dd.encode('ascii', 'ignore')
            add.append(dd)
            add_list=[]
        x=0
        while x<len(add):
            tmp_add =str(add[x])+" "+str(add[x+1])
            add_list.append(tmp_add)
            x+=2
        return add_list
        
    def visualize(price,address):
        df=pd.DataFrame({'price':price,'address':address})
        return df
    
    price=find_the_all_price()
    address=find_the_address()
    
    dataframe=visualize(price,address)
    return dataframe    
```

```python
main()
```
